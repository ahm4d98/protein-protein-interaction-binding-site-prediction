```python
import os 
import pandas as pd
%load_ext tensorboard
import datetime
import tensorflow as tf
import datetime

l=[]
j=os.listdir(r"C:\Users\ahmed hatem\Downloads\Ahmed_Hatem20178003-mohamed khaled-20178056-mohamed-ali-20178036\ready for models 0.9")
for i in range(0,len(j)):
    os.chdir(r"C:\Users\ahmed hatem\Downloads\Ahmed_Hatem20178003-mohamed khaled-20178056-mohamed-ali-20178036\ready for models 0.9")
    df = pd.read_csv(j[i])
    col=df.columns
    df.reset_index(drop=True, inplace=True)
    l.append(df)
    df.dropna( axis=0, inplace=True)
df=pd.concat(l,axis=0,ignore_index=True)
```

```python
df=df.drop(['X1','Unnamed: 2','Y  V','Y.1','V.1'],axis=1)
```

```python
df.dropna( axis=0, inplace=True)
```

```python
x = df.iloc[:, 0 :170]
y = df['prediction']
```

```python
x
```

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten, MaxPooling1D,Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import os
```

```python
model = Sequential()
model.add(Conv1D(32,1,padding="same",activation="relu", input_shape=(170,1)))
model.add(MaxPooling1D((2)))
model.add(Conv1D(8,1,padding="same",activation="softmax"))
model.add(MaxPooling1D((2)))

model.add(Dense(16, activation="softmax"))

model.add(Flatten())

model.add(Dense(2, activation = 'softmax'))
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = "adam",metrics = ['accuracy'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
x_train.shape
```

```python
x_train=np.expand_dims(x_train, axis=2)
```

```python
x_test=np.expand_dims(x_test, axis=2)
```

```python
from tensorflow.keras.callbacks import TensorBoard

tensorboard = TensorBoard(log_dir='.\logs',histogram_freq=1,write_images=True)

keras_callbacks = [tensorboard]

history=model.fit(x_train, y_train, batch_size=64,epochs=100, verbose=0,validation_data=(x_test, y_test),callbacks=keras_callbacks)

acc = model.evaluate(x_train, y_train)
print("Loss:", acc[0], "Accuracy:", acc[1])

pred = model.predict(x_test)
pred_y = pred.argmax(axis=-1)
pred_y
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred_y)
accuracy
```

```python
import matplotlib.pyplot as plt 
from sklearn import metrics 
plt.figure(dpi=1500)
prob_train = np.squeeze(model.predict_proba(x_train)[:,1].reshape(1,-1))
prob_test = np.squeeze(model.predict_proba(x_test)[:,1].reshape(1,-1))
 # false positive rate, true positive rate, thresholds
fpr1, tpr1, thresholds1 = metrics.roc_curve(y_test, prob_test)
fpr2, tpr2, thresholds2 = metrics.roc_curve(y_train, prob_train)
 # auc score
auc1 = metrics.auc(fpr1, tpr1)
auc2 = metrics.auc(fpr2, tpr2)
plt.figure(figsize=(8,8))
 # plot auc 
plt.plot(fpr1, tpr1, color='blue', label='Test ROC curve area = %0.2f'%auc1)
plt.plot(fpr2, tpr2, color='green', label='Train ROC curve area = %0.2f'%auc2)
plt.plot([0,1],[0,1], 'r--')
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('False Positive Rate', size=14)
plt.ylabel('True Positive Rate', size=14)
plt.legend(loc='lower right')
plt.savefig("ROC.png",dpi=1500)

plt.show() 
```

```python
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
kappa = cohen_kappa_score(y_test, pred_y)
print('Cohens kappa: %f' % kappa)
# ROC AUC
auc = roc_auc_score(y_test, pred_y)
print('ROC AUC: %f' % auc)
# confusion matrix
matrix = confusion_matrix(y_test, pred_y)
print(matrix)
```
